```python
def walking(origin,destination,sig=None)->dict:
    url = 'https://restapi.amap.com/v3/direction/walking?parameters'
    params={
        'key':"9f7b177f4227578e1842731c2bb198e6",
        'origin':'113.258953,23.114944',
        'destination':'113.260032,23.124473',
        'output':'json'
    }
    response = requests.get(url,params=params)
    data = response.json()
    return data
```

```python
from PIL import Image
import io
from io import BytesIO
f = BytesIO()
def staticmap(location,zoom,size=None,scale=1,markers=None,labels=None,paths=None,traffic=0,page=None,sig=None)->dict:
    url = 'https://restapi.amap.com/v3/staticmap?parameters'
    params={
        'key':'9f7b177f4227578e1842731c2bb198e',
        'location':location,
        'zoom':zoom,
        'size':size,
        'scale':scale,
        'markers':markers,
        'labels':labels,
        'paths':paths,
        'traffic':traffic,
        'sig':sig,
        'output':'json'
    }
    response = requests.get(url,params=params)
    data = Image.open(f(response.content))
    return data
```
